```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Binary Step Activation Function
def binary_step(x):
    return np.heaviside(x, 1)

x = np.linspace(-10, 10)
plt.plot(x, binary_step(x))
plt.title('Activation Function: Binary Step')
plt.show()
```

```python
# Linear Activation Function

def linear(x):
    y = x + 1
    return y

x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.title('Linear Activation Function')
plt.show()
```

```python
# Sigmoid Activation Function

def sigmoid(x):
    e = 2.17
    sigmoid_x = 1 / (1 + e**(-x))
    return sigmoid_x

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
# Tanh Activation Function

def tanh(x):
    f_of_x = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
    return f_of_x

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.title('Tanh Activation Function')
plt.show()
```

```python
# ReLU Activation Function

def relu(x):
    x_val = []
    for i in x:
        if i > 0:
            x_val.append(i)
        else:
            x_val.append(0)
    return x_val

x = np.linspace(-10, 10)
plt.plot(x, relu(x))
plt.title('ReLU Activation Function')
plt.show()
```

```python
# Softmax Activation Function

def softmax(x):
    sft_list = []
    for ele in x:
        b = np.exp(ele)
        sft_list.append(b)
    summation = sum(sft_list)

    softmax_x = np.exp(x)/summation

    return softmax_x

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.title('Softmax Activation Function')
plt.show()
```
