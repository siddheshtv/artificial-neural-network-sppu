```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
num_points = 100
points = np.random.rand(num_points, 2) * 10
labels = np.array([1 if p[0] + p[1] > 10 else 0 for p in points])


weights = np.random.rand(2)
bias = np.random.rand()


learning_rate = 0.1
epochs = 100

for _ in range(epochs):
    for point, label in zip(points, labels):
        activation = np.dot(point, weights) + bias
        prediction = 1 if activation >= 0 else 0
        error = label - prediction
        weights += learning_rate * error * point
        bias += learning_rate * error


plt.figure(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.xlabel('X1')
plt.ylabel('X2')


x_values = np.linspace(0, 10, 100)
y_values = -(weights[0] * x_values + bias) / weights[1]
plt.plot(x_values, y_values, color='red', linestyle='dashed', label='Decision Boundary')

plt.legend()
plt.title('Perceptron Learning Rule - Decision Regions')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```
